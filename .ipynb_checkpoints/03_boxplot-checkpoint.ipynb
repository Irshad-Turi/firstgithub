```python
#Plot for single variable
#import library
import seaborn as sns
import pandas as pd
import numpy
import matplotlib.pyplot as plt
#canvas  (Baloon board)
sns.set(style="whitegrid")
#
tip = seaborn.load_dataset("tips")
my_pal={"Female":"#B10BC5","Male":"#62C50B"}  #using own optional color for each bar
sns.boxplot(x="day",
               y="tip",
               data=tip, saturation=1, hue="sex",palette=my_pal  )
```
